# Analyse des résultats

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = '/home/robin/Documents/DGS/Livraison_L3/analyse _serge.xlsx'
df_topics = pd.read_excel(path,sheet_name='thèmes')
df_clusters = pd.read_excel(path,sheet_name = 'clusters')

path_2 = '/home/robin/Documents/DGS/Livraison_L3_12_10_2020/analyse_12_10_2020.xlsx'
df_topics_2 = pd.read_excel(path_2,sheet_name='thèmes')
df_clusters_2 = pd.read_excel(path_2,sheet_name = 'clusters')

path_4 = '/home/robin/Documents/DGS/Annexe/L3/14_10_2020/analyse_serge.xlsx'
df_topics_4 = pd.read_excel(path_4,sheet_name='thèmes')
df_clusters_4 = pd.read_excel(path_4,sheet_name = 'clusters')
```

## 1. Analyse des thèmes

```python
df_topics.groupby('score').count()['Numero'].map(lambda x: x/len(df_topics)).plot(kind='barh',figsize=(10,5))
plt.title(f"Distribution des scores pour les topics (n = {len(df_topics)})")
plt.show()


#df_topics_2.groupby('score').count()['Numero'].map(lambda x: x/len(df_topics_2)).plot(kind='barh',figsize=(10,5))
#plt.title(f"Distribution des scores pour les topics (n = {len(df_topics_2)})")
#plt.show()

df_topics_4.groupby('score').count()['Numero'].map(lambda x: x/len(df_topics_4)).plot(kind='barh',figsize=(10,5))
plt.title(f"Distribution des scores pour les topics (n = {len(df_topics_4)})")
plt.show()
```

```python
n_2 = len(df_topics[df_topics['score']==2])
df_topics = df_topics[df_topics['score']==2]
df_topics.groupby('titre').count()['score'].map(lambda x: x/n_2).plot(kind='barh',figsize=(10,5))
plt.title(f"Distribution des titres pour les topics pertinents (n = {n_2})")
plt.show()

"""n_2 = len(df_topics_2[df_topics_2['score']==2])
df_topics_2 = df_topics_2[df_topics_2['score']==2]
df_topics_2.groupby('titre').count()['score'].map(lambda x: x/n_2).plot(kind='barh',figsize=(10,5))
plt.title(f"Distribution des titres pour les topics pertinents (n = {n_2})")
plt.show()"""

n_4 = len(df_topics_4[df_topics_4['score']==2])
df_topics_4 = df_topics_4[df_topics_4['score']==2]
df_topics_4.groupby('titre').count()['score'].map(lambda x: x/n_2).plot(kind='barh',figsize=(10,5))
plt.title(f"Distribution des titres pour les topics pertinents (n = {n_2})")
plt.show()
```

## 2 Analyse des clusters

```python
df_clusters.groupby('score').count()['Numero'].map(lambda x: x/len(df_clusters)).plot(kind='barh',figsize=(10,5))
plt.title(f"Distribution des scores pour les clusters (n = {len(df_clusters)})")
plt.show()

df_clusters_2.groupby('score').count()['Numero'].map(lambda x: x/len(df_clusters_2)).plot(kind='barh',figsize=(10,5))
plt.title(f"Distribution des scores pour les clusters (n = {len(df_clusters_2)})")
plt.show()

df_clusters_4.groupby('score').count()['Numero'].map(lambda x: x/len(df_clusters_4)).plot(kind='barh',figsize=(10,5))
plt.title(f"Distribution des scores pour les clusters (n = {len(df_clusters_4)})")
plt.show()
```

```python
n_2 = len(df_clusters[df_clusters['score']==2])
df_clusters.groupby('titre').count()['score'].map(lambda x: x/n_2).sort_values().plot(kind='barh',figsize=(10,5))
plt.title(f"Distribution des titres pour les clusters pertinents (n = {n_2})")
plt.show()

n_2 = len(df_clusters_2[df_clusters_2['score']==2])
df_clusters_2.groupby('titre').count()['score'].map(lambda x: x/n_2).sort_values().plot(kind='barh',figsize=(10,5))
plt.title(f"Distribution des titres pour les clusters pertinents (n = {n_2})")
plt.show()


n4 = len(df_clusters_4[df_clusters_4['score']==2])
df_clusters_4.groupby('titre').count()['score'].map(lambda x: x/n_4).sort_values().plot(kind='barh',figsize=(10,5))
plt.title(f"Distribution des titres pour les clusters pertinents (n = {n_4})")
plt.show()
```
